#### Outliers for average rating using IQR method

```python
Q1 = np.percentile(books_df['average_rating'], 25, interpolation = 'midpoint')
Q3 = np.percentile(books_df['average_rating'], 75, interpolation = 'midpoint')
IQR = Q3 - Q1

print('Q1= ',Q1)
print('Q3= ',Q3)
print ('IQR= ',IQR)

print('Q3+1.5*IQR= ',Q3+1.5*IQR)
print('Q1-1.5*IQR= ',Q1-1.5*IQR)


print("Old Shape: ", books_df.shape)
```

```python
books_df.head(10)
```

```python
books_df.average_rating.count()
```

```python
#upper bound
upper = np.where(books_df['average_rating'] >= (Q3+1.5*IQR))
print('Number of "high" outliers: ',len(upper[0]))

# Lower bound
lower = np.where(books_df['average_rating'] <= (Q1-1.5*IQR)) 
print('Number of "low" outliers: ',len(lower[0]))

print('Total number of outliers:', len(lower[0])+ len(upper[0]),)

print('Total before and after dropping outliers:', books_df.average_rating.count(), ' / ',books_df.average_rating.count()-(len(lower[0])+ len(upper[0])),'\n')


print('Ratings above upper bound are: ', upper,'\n')#return a tuple containing rows with average rating >4.682 - row starting at 0
print('Ratings below lower are: ', lower) #return a tuple containing rows with average rating < 3.2225 - row starting at 0
```
